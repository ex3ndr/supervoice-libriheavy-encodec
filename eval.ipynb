```python
import torch
from IPython.display import Audio, display
from vocos import Vocos
```

```python
vocos = Vocos.from_pretrained("charactr/vocos-encodec-24khz")
```

```python
audio_tokens = torch.load("./encoded_datasets/librilight/100/sea_fairies_0812_librivox_64kb_mp3/01_baum_sea_fairies_64kb_0.pt")
features = vocos.codes_to_features(audio_tokens)
bandwidth_id = torch.tensor([2])  # 6 kbps
audio = vocos.decode(features, bandwidth_id=bandwidth_id)
display(Audio(data=audio.cpu(), rate=24000))
print(audio.shape)
print(audio_tokens.shape)
```
